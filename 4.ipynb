```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder

sns.set()


kmeans = KMeans(n_clusters=3, n_init=10)
df1 = pd.read_csv("C:\\Users\\User\\Downloads\\Iris.csv")

X = df1[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]

model = KMeans(n_clusters=3)
model.fit(X)
df1['cluster'] = model.predict(X)

plt.scatter(df1['SepalLengthCm'], df1['SepalWidthCm'], c=df1['cluster'], s=30*(3*df1['cluster']+1), marker="v")


wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wcss)
plt.title('Выбор количества кластеров методом локтя')
plt.xlabel('Количество кластеров')
plt.ylabel('WCSS')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder

sns.set()
kmeans = KMeans(n_clusters=3, n_init=10)


# Загружаем данные
data = pd.read_csv("C:\\Users\\User\\Downloads\\insurance.csv")
# Преобразуем категориальные переменные "sex" и "region" в числовой формат
data = pd.get_dummies(data, columns=['sex', 'region'], drop_first=True)
# Удаляем переменные, определяющие класс (smoker)
data_without_labels = data.drop(columns=['smoker'])
# Применяем метод K-средних
kmeans = KMeans(n_clusters=3, random_state=42)
data['cluster'] = kmeans.fit_predict(data_without_labels)
# Строим график
plt.scatter(data['age'], data['bmi'], c=data['cluster'], cmap='viridis', s=50, alpha=0.5)
plt.xlabel('Age')
plt.ylabel('BMI')
plt.title('Clusters')
plt.show()
# Создаем таблицу сопряженности
contingency_table = pd.crosstab(data['sex_male'], data['cluster'])
print(contingency_table)
# Определяем оптимальное количество кластеров с помощью метода локтя
wcss = []
for i in range(1, 11):
    #настроим параментры модели
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=42)
    #обучим модель данных с разным набором кластеров
    kmeans.fit(data[['age', 'bmi']])
    #для каждого кластера рассчитаем ошибку и поместим в список
    wcss.append(kmeans.inertia_)
#передадим последовательность кластеров и ошибку wcss
plt.plot(range(1, 11), wcss)
plt.title('Number of Clusters using Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```
